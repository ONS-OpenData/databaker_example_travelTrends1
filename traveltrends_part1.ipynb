```python
"""
First we import databaker and load the specific tabs into memory.
"""

from databaker.framework import *

# name the input and output file
inputfile = 'travelTrends4.xls'
outputfile ='output_travelTrends4.csv'

# load the tabs we want
tabsWeWant = ['4.12a', '4.12b']
tabs = loadxlstabs(inputfile, tabsWeWant)

"""
Secondly we start a loop so we can look at each of the selected tabs
"""

conversionsegments = [] # we'll need this later
for tab in tabs:
    
    """Now the cell selection stage"""
    
    # First, we're going to select the business and holidays cells from row 6 as a variable typeOfTravel
    holiday = tab.excel_ref("6").filter(contains_string("Holiday"))
    business = tab.excel_ref("6").filter(contains_string("Business"))
    typeOfTravel = holiday | business

    # Now we'll select the area, removing any Totals
    allAreas = tab.excel_ref('A').is_not_blank().is_not_whitespace()
    unwantedAreas = tab.excel_ref('A').filter(contains_string("Total"))
    area = allAreas - unwantedAreas
    
    # And now we select the observations with a waffle
    observations = typeOfTravel.waffle(area).is_not_blank().is_not_whitespace()
    
    
    """
    Now we need to establish the relation between our observations
    and our dimensions (typeOfTravel, area, time)
    
    I've included a colored image below that should help with working through the logic of what this is doing.
    The key thing to remember is we're defining relationships of each dimension - RELATIVE TO THE OBSERVATIONS
    """
    dimensions = [
              HDim(area, GEOG, DIRECTLY, LEFT), 
              HDim(typeOfTravel, "Type of Travel", DIRECTLY, ABOVE), 
              HDimConst(TIME, "2016")    # the time for all is 2016, so we just set it as constant 
                 ]
    
    """
    Now we pass the info we need to flatten our selection to the databaker function 'ConversionSegment'
    Then add it to a list.
    """
    conversionsegment = ConversionSegment(tab, dimensions, observations) # < --- processing
    conversionsegments.append(conversionsegment) # <-- adding result of processing this tab to our list
    
# Write every 'ConversionSegment' in the list to a csv
writetechnicalCSV(outputfile, conversionsegments)

```
